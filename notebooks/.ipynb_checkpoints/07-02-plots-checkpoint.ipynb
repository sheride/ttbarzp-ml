```python
def getFeatureImportance(importances, num=10, toPrint=True):
    max_col = max([len(name) for name in DA.cols])
    indices = np.argsort(-np.abs(importances))
    feature_dict = {feature: imp for feature, imp in zip(DA.cols, importances)}
    if toPrint:
        for i in indices[:num]:
            feature = DA.cols[i]
            val = feature_dict[feature]
            sign = '+' if val >= 0 else '-'
            print('{0:{1}} {3}{2:.4f}'.format(feature + ':', max_col + 3, np.abs(val), sign))
    return [feature_dict, indices]
```

```python
def get_sig_bg_probs(model):
    sig_test = model.test_pred[(model.test_label == 1)]
    bg_test = model.test_pred[(model.test_label == 0)]
    return [model.model.predict_proba(sig_test)[:,1], 
            model.model.predict_proba(bg_test)[:,1]]
```

```python
def plot_mass_sens(min_tpr=0.3, min_fpr=10**-6):
    # useful constants
    PI = PhysicsInfo()
    sig = 5
    bg_cs = PI.bg_cs()
    lumi = 3000 # fb
    bg_yield = PI.conv * bg_cs * lumi

    # sampled TPRs, FPRs
    tprs = np.linspace(min_tpr, 1, 100)
    fprs = np.sort(np.logspace(np.log10(min_fpr), np.log10(0.6), 100))

    # computing maximal sensitive mass as a function of tpr, fpr 
    coef = [[[-(PI.conv * tpr * lumi)**2, sig**2 * PI.conv * tpr * lumi, sig**2 * fpr * bg_yield] for fpr in fprs] for tpr in tprs]
    sig_cs = [[np.amax(np.roots(elem)) for elem in row] for row in coef]
    masses = PI.Zp_mass(sig_cs)

    # preparing parameters for plotting
    xs = tprs
    ys = np.log10(fprs)
    data = masses
    res = 1000
    cvalues = [500, 750, 1000, 1500, 2000]
    clabels = [str(a) for a in cvalues]
    manual = [(0.85, -1), (0.85, -2), (0.85, -3), (0.85, -3.5), (0.85, -6)]

    # plotting
    ext = [xs[0], xs[-1], ys[0], ys[-1]]
    pts = np.array([[x,y] for x in xs for y in ys])
    data = np.array(data).flatten()
    grid = np.array(
            [[[x,y] for x in np.linspace(ext[0], ext[1], res)]
            for y in np.linspace(ext[2], ext[3], res)])
    interp = scipy.interpolate.griddata(pts, data, grid)
    with plt.rc_context(settings):
        plt.imshow(interp, origin='lower', aspect='auto', extent=ext, cmap='winter')
        cbar = plt.colorbar()
        plt.xlabel(r'{}'.format('True Positive Rate'))
        plt.ylabel(r'{}'.format('False Postive Rate'))
        cbar.set_label(r'{}'.format("Z' Mass Sensitivity (GeV)"), rotation=90)
        result = plt.contour(grid[:,:,0], grid[:,:,1], interp, cvalues, colors='white',
                                     linewidths=3, linestyles='dashed')
        fmt = {lev:lab for lev, lab in zip(result.levels, clabels)}
        plt.clabel(result, result.levels, inline=True, fmt=fmt, fontsize=32,
                   manual=manual)
        plt.yticks([-1, -2, -3, -4, -5, -6], 
                   [r'$10^{-1}$', r'$10^{-2}$', r'$10^{-3}$', r'$10^{-4}$', '$10^{-5}$', '$10^{-6}$'])
        plt.xlim(min_tpr, 1)
        plt.ylim(np.log10(min_fpr), np.log10(0.6))
```
